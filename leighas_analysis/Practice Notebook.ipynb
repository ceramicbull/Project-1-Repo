```python
# import dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# read in raw data of prime rates using pandas
prime_rates_data_to_load = 'Resources/lr _prime_rates.csv'
lr_prime_rates = pd.read_csv(prime_rates_data_to_load)
lr_prime_rates
```

```python
# read in raw data of mortgage rates using pandas
mortgage_rates_data_to_load = 'Resources/lr_mortgage_rates.csv'
lr_mortgage_rates = pd.read_csv(mortgage_rates_data_to_load)
lr_mortgage_rates
```

```python
# plot bank prime rates vs mortgage rates
mortgage_rates_to_plot = lr_mortgage_rates['MORTGAGE30US']
prime_rates_to_plot = lr_prime_rates['MPRIME']

years = lr_mortgage_rates['DATE']

plt.figure(figsize=(20,15))
prime_loan_plot, = plt.plot(years, prime_rates_to_plot, color="blue", label="Prime Rates" )

mortgage_plot, = plt.plot(years, mortgage_rates_to_plot, color='green', label="Mortgage Rates")

plt.xticks(rotation="vertical")
#plt.legend(handles=[prime_rates_to_plot, mortgage_rates_to_plot], loc="best")

plt.show()
```
